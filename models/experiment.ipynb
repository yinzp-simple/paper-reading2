```python
import numpy as np
import matplotlib.pyplot as plt
def draw_loss(filepath):
    import matplotlib.pyplot as plt
    loss_list = list(np.load(filepath))
    plt.plot(list(range(1, len(loss_list) + 1)), loss_list)
```
